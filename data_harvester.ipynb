```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from nltk.tokenize import sent_tokenize
```

```python
# creating list for each language

eng = []
mal = []
hin = []
tam = []
por = []
fre = []
ger = []
dut = []
spa = []
gre = []
rus = []
dan = []
ita = []
tur = []
swe = []
ara = []
jap = []
kor = []
cat = []
nor = []

# creating a nested list
data_list = [eng, mal, hin, tam, por, fre, ger, dut, spa, gre, rus, dan, ita, tur, swe, ara, jap, kor, cat, nor]

# language notation for wikipedia search
language = ["en","ml","hi","ta","pt","fr", "de", "nl","es","el","ru","da","it","tr","sv","ar", "jp", "ko", "ca", "no"]
```

```python
def text_scrap(topic):
    
    if topic is None:
        topic = "Special:Random"
        for i in range(len(language)):
            paragraphs = []
            while len(paragraphs) < 10:
                url = f"https://{language[i]}.wikipedia.org/wiki/{topic}"

                # retrieving the data from the above url source
                request = requests.get(url)
                soup = BeautifulSoup(request.text, "lxml")

                # finding all p tag in the above url page
                paragraphs = soup.find_all("p")
            
            print(f"{language[i]}: {len(paragraphs)}")

            for paragraph in paragraphs:

                # appending each paragraph
                data_list[i].append(paragraph.text.strip())
        return

    
    for i in range(len(language)):
        
        # url for wikipedia search
        url = f"https://{language[i]}.wikipedia.org/wiki/{topic}"
        print(url)
        
        # retrieving the data from the above url source
        request = requests.get(url)
        soup = BeautifulSoup(request.text, "lxml")
        
        # finding all p tag in the above url page
        paragraphs = soup.find_all("p")
        
        for paragraph in paragraphs:
            
            # appending each paragraph
            data_list[i].append(paragraph.text.strip())
```

```python
text_data = []
for text_list in data_list:
    text_list = " ".join(text_list)
    text_data.append(text_list)
```

```python
def dataframe(text, lang):
    tokenized_text = sent_tokenize(text)
    text_df = pd.DataFrame({
        "Text":tokenized_text,
        "Language": lang
    })
    
    return text_df
    
```

```python
df0 = dataframe(text_data[0], "English")
df1 = dataframe(text_data[1], "Malayalam")
df2 = dataframe(text_data[2], "Hindi")
df3 = dataframe(text_data[3], "Tamil")
df4 = dataframe(text_data[4], "Portuguese")
df5 = dataframe(text_data[5], "French")
df6 = dataframe(text_data[6], "German")
df7 = dataframe(text_data[7], "Dutch")
df8 = dataframe(text_data[8], "Spanish")
df9 = dataframe(text_data[9], "Greek")
df10 = dataframe(text_data[10], "Russian")
df11 = dataframe(text_data[11], "Danish")
df12 = dataframe(text_data[12], "Italian")
df13 = dataframe(text_data[13], "Turkish")
df14 = dataframe(text_data[14], "Swedish")
df15 = dataframe(text_data[15], "Arabic")
df16 = dataframe(text_data[16], "Japanese")
df17 = dataframe(text_data[17], "Korean")
df18 = dataframe(text_data[18], "Catalan")
df19 = dataframe(text_data[19], "Norwegian")
```
